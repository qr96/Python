```python
import cloudscraper
from bs4 import BeautifulSoup

#name은 검색어, num은 갯수
def parseImg(name, num):
    
    url = 'https://pixabay.com/ko/images/search/'+name+'/?pagi='
    scraper = cloudscraper.create_scraper()
    hdr = {'User-Agent': 'Mozilla/5.0'}
    pageIdx=1 #n번째 페이지도 받음
    
    while True:
        req = scraper.get(url+str(pageIdx),headers = hdr)
        source = req.text
        soup = BeautifulSoup(source, "html.parser")
        tagArr = soup.find_all("img") #img가 들어간 태그 모두 찾음
        tagNum  = len(tagArr) #태그들의 갯수
        
        if tagArr is None:
            break
        elif tagNum > num:
            del tagArr[num:] #tag 데이터의 크기 조절

        for tag in tagArr:
            feature = tag.get('alt')
            srclink = tag.get('srcset')
            if srclink is None:
                srclink = tag.get('data-lazy')
            if feature:
                print("관계단어 : "+feature)
                print("이미지 링크 : "+srclink)
                print("\n")
        
        if tagNum < num:
            num-=tagNum
            pageIdx+=1
        else:
            break
            
            
```

```python
parseImg("apple", 820)
```
