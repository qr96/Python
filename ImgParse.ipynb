```python
import cloudscraper
from bs4 import BeautifulSoup

#name은 검색어, num은 갯수
def parseImg(name, num):
    url = 'https://pixabay.com/ko/images/search/'+name
    scraper = cloudscraper.create_scraper()
    
    
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = scraper.get(url,headers = hdr)
    source = req.text
    soup = BeautifulSoup(source, "html.parser")
    
    for link in soup.findAll("img"):
        print(link)

```

```python
parseImg("apple", 5)
```
